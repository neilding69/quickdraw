```python
from utils.read_data import get_dataset
train_R,train_X,train_Y,test_R,test_X,test_Y=get_dataset(dataset_name='1102_05b633244')
X=train_X+test_X
Y=train_Y+test_Y
R=train_R+test_R
```

We select those samples to show the paintings

```python
sample=[0,1,3,7,9]
```

```python
def draw_pictures(X,Y):
    words = Y
    drawings = X
    import matplotlib.pyplot as plt
    plt.figure(figsize=(10, 10))
    for i, drawing in enumerate(drawings):
        plt.subplot(330 + (i+1))
        for x,y in drawing:
            plt.plot(x, y, marker='.')
            plt.tight_layout()
            plt.title(words[i]);
            plt.axis('off')
```

```python
import numpy as np

nX=np.array(X)[sample]
nY=np.array(Y)[sample]
nR=np.array(R)[sample]
d={0:'penguin',1:'calendar',2:'blackberry',3:'teddy-bear',4:'snowman',}
nY=[d[i] for i in nY]
```

```python
draw_pictures(nX,nY)
```

Then we show the distribution of the number of stroke (first plot) and the maximal length of stroke (second plot) in a painting

```python
def count(l):
    x=list(range(0,max(l)))
    y=[l.count(i) for i in x]
    return x,y
def get_max_len_stroke(x):
    l_len=[]
    l_stroke=[]
    for i,item in enumerate(x):
        l_stroke.append(len(item))
        max_len=0
        for j,stroke in enumerate(item):
            max_len=max(max_len,len(stroke[0])) 
        l_len.append(max_len)
    return l_len,l_stroke
l_len,l_strock=get_max_len_stroke(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x,y=count(l_strock)
plt.figure()
plt.plot(x,y)
plt.show()
x,y=count(l_len)
plt.figure()
plt.plot(x,y)
plt.show()
```
